10) Selecciona el maximo de datos faltantes. Calcula que porcentaje de los datos representan los datos faltantes.

```python
import numpy as np
```

```python
# Obtengo el maximo valor de datos faltantes y lo guardo en la variable masnan
maxnan_np = np.max(expvida.isnull().sum())
print(maxnan_np)

maxnan_pd = expvida.isnull().sum().max()
print(maxnan_pd)
```

```python
datos_no_faltantes = expvida.notnull().values.sum()
datos_faltantes = expvida.isnull().values.sum()
total_datos = datos_no_faltantes + datos_faltantes
print(total_datos)

total = expvida.shape[0] * expvida.shape[1]
print(total)

porcentaje = (datos_faltantes/ total_datos) * 100
print(porcentaje)
```

```python
# Obtengo la cantidad de filas total y divido el maximo valor de datos faltantes por este numero
# y lo multiplico por 100
total_filas = expvida.shape[0]
porc = maxnan_np/total_filas * 100
```

```python
# Imprimo el maximo porcentaje de datos faltantes. Uso f-strings.
print(f"El porcentaje del máximo valor de datos faltantes es: {porc:.2f}")
```

13) ¿Cual es el promedio de expectativa de vida (Life expectancy) por país?

```python
# Obtengo el promedio por pais
#media_pais = expvida.grouphy ()
media_pais = expvida.groupby('Country')[['life_expectancy']].mean()
media_pais
```

14) ¿Cual es la expectativa de vida mas alta y mas baja? Usa la funcion de pandas y la funcion de NumPy y compara el resultado. ¿Que paises tienen esas expectativas de vida?

```python
# Maxima expectativa de vida usando numpy
pais_max_exp = np.max(media_pais)
```

```python
# Maxima expectativa de vida usando pandas
pais_max_expe = media_pais.max()
```

```python
# Minima expectativa de vida usando numpy
pais_min_exp = np.min(media_pais)
```

```python
# Minima expectativa de vida usando pandas
pais_min_expe = media_pais.min()
```

```python
# Muestro los resultados maximos
print(f'Máxima expectativa de vida usando Numpy: {pais_max_exp}')
print(f'Máxima expectativa de vida usando Pandas: {pais_max_expe}')
```

```python
# Muestro los resultados minimos
print(f'Mínima expectativa de vida usando Numpy: {pais_min_exp}')
print(f'Mínima expectativa de vida usando Pandas: {pais_min_expe}')
```

```python
# Filtro las columnas para mostrar solo las filas que tengan valor
# igual al maximo
result_max = media_pais[media_pais['life_expectancy'] == pais_max_exp]
result_max
```

18) ¿Cuantos paises presentaron algun caso de Sarampion (**Measles**) en el 2015? **_Pista_**: Lee acerca de la función [.count()](http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.count.html?highlight=count#pandas.DataFrame.count)

```python
# Cuento cuantas filas tienen valores distintos de cero
data = expvida[expvida['measles'] != 0]
cant_filas = data['measles'].count()
print(f'La cantidad de filas que tienen valores distintos de ceros en la columna "measles" es: {cant_filas}')
```

```python
data_filtered = expvida[(expvida['Year'] == 2015) & (expvida['measles'] > 0)]
data_filtered = data_filtered.drop_duplicates(subset='Country')  #Me aseguro que no haya un país duplicado
data_filtered
```

```python
resultado = data_filtered.shape[0]
print(f'La cantidad de países que presentaron algún caso de sarampión (measles) en el 2015 es: {resultado}')
```
